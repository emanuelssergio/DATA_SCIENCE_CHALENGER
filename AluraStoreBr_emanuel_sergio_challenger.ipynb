### Importação dos dados



````python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
````

#1. Análise do faturamento


````python

# loja 1
preco_loja1=loja['Preço']
faturamento_loja1=sum(preco_loja1)
faturamento_loja1=round(faturamento_loja1,2)
media_por_produto_loja1=faturamento_loja1/len(preco_loja1)
print(f'faturamento da primeira loja: {faturamento_loja1}')
print(f'media de preço por produto:   {round(media_por_produto_loja1,2)}')
print('')


# loja 2
preco_loja2=loja2['Preço']
faturamento_loja2=sum(preco_loja2)
faturamento_loja2=round(faturamento_loja2,2)
media_por_produto_loja2=faturamento_loja2/len(preco_loja2)
print(f'faturamento da segunda loja:  {faturamento_loja2}')
print(f'media de preço por produto:   {round(media_por_produto_loja2,2)}')
print('')

#loja 3
preco_loja3=loja3['Preço']
faturamento_loja3=sum(preco_loja3)
faturamento_loja3=round(faturamento_loja3,2)
media_por_produto_loja3=faturamento_loja3/len(preco_loja3)
print(f'faturamento da terceira loja: {faturamento_loja3}')
print(f'media de preço por produto:   {round(media_por_produto_loja3,2)}')
print('')

#loja 4

preco_loja4=loja4['Preço']
faturamento_loja4=sum(preco_loja4)
faturamento_loja4=round(faturamento_loja4,2)
media_por_produto_loja4=faturamento_loja4/len(preco_loja4)
print(f'faturamento da quarta loja:   {faturamento_loja4}')
print(f'media de preço por produto:   {round(media_por_produto_loja4,2)}')
print('')




#maior faturamento
faturamentos = [faturamento_loja1,faturamento_loja2,faturamento_loja3,faturamento_loja4]
maior_faturamento=max(faturamentos)
loja_maior_faturamento=faturamentos.index(maior_faturamento)+1
print(f'a loja com maior faturamento é a loja{loja_maior_faturamento} com o faturameto de {maior_faturamento} reais')







````

# 2. Vendas por Categoria


````python
import matplotlib.pyplot as plt
#loja 1
categorias_loja_1 = loja['Categoria do Produto']
categorias=['utilidades domesticas', 'esporte e lazer', 'brinquedos', 'instrumentos musicais', 'eletrodomesticos', 'eletronicos', 'moveis', 'livros']

print(categorias)
#todas as lojas tem as mesmas categorias


categoria=list(zip(loja['Categoria do Produto']))

produdos=list(zip(loja['Categoria do Produto'],loja['Preço']))

preços_brinquedos=[]
preços_eletronicos=[]
preços_utilidades_domesticas=[]
preços_instrumentos_musicais=[]
preços_eletrodomesticos=[]
preços_moveis=[]
preços_esporte_lazer=[]
preços_livros=[]

for categoria,preco in produdos:
  if categoria == categorias[0]:
    preços_utilidades_domesticas.append(preco)

  elif categoria == categorias[1]:
    preços_esporte_lazer.append(preco)

  elif categoria == categorias[2]:
    preços_brinquedos.append(preco)

  elif categoria == categorias[3]:
    preços_instrumentos_musicais.append(preco)

  elif categoria == categorias[4]:
    preços_eletrodomesticos.append(preco)

  elif categoria == categorias[5]:
    preços_eletronicos.append(preco)

  elif categoria ==categorias[6]:
    preços_moveis.append(preco)

  elif categoria == categorias[7]:
    preços_livros.append(preco)



quantidade_domesticas=(len(preços_utilidades_domesticas))
quantidade_esporte_lazer=(len(preços_esporte_lazer))
quantidade_brinquedos=(len(preços_brinquedos))
quantidade_musicais=(len(preços_instrumentos_musicais))
quantidade_eletrodomesticos=(len(preços_eletrodomesticos))
quantidade_eletronicos=(len(preços_eletronicos))
quantidade_moveis=(len(preços_moveis))
quantidade_livros=(len(preços_livros))
quantidades=[quantidade_domesticas,quantidade_esporte_lazer,quantidade_brinquedos,quantidade_musicais,quantidade_eletrodomesticos,quantidade_eletronicos,quantidade_moveis,quantidade_livros]
print(quantidades)

#protudo mais vendido

maior_valor = max(quantidades)
mais_vendido=int(quantidades.index(maior_valor))
print(f'o produto mais vendido da loja 1 é {categorias[mais_vendido]} com {maior_valor} unidades vendidas')
print('\n')

#protudo menos vendido

menor_valor = min(quantidades)
menos_vendido=int(quantidades.index(menor_valor ))
print(f'o produto menos vendido da loja 1 é {categorias[menos_vendido]} com {menor_valor} unidades vendidas')
print('\n')


#faturamento por categoria
print(' o faturamento por categoria é:')
loja1_brinquedos=sum(preços_brinquedos)
loja1_eletronicos=sum(preços_eletronicos)
loja1_utilidades_domesticas=sum(preços_utilidades_domesticas)
loja1_instrumentos_musicais=sum(preços_instrumentos_musicais)
loja1_eletrodomesticos=sum(preços_eletrodomesticos)
loja1_moveis=sum(preços_moveis)
loja1_esporte_lazer=sum(preços_esporte_lazer)
loja1_livros=sum(preços_livros)
print(f' faturamento da loja 1 na cadegoria de brinquedos:             {loja1_brinquedos} reais')
print(f' faturamento da loja 1 na cadegoria de eletronicos:           {round(loja1_eletronicos,2)} reais')
print(f' faturamento da loja 1 na cadegoria de utilidades_domesticas:  {round(loja1_utilidades_domesticas,2)} reais')
print(f' faturamento da loja 1 na cadegoria de instrumentos_musicais: {round(loja1_instrumentos_musicais,2) } reais')
print(f' faturamento da loja 1 na cadegoria de eletrodomesticos:      {round(loja1_eletrodomesticos,2)} reais')
print(f' faturamento da loja 1 na cadegoria de moveis:                {round(loja1_moveis,2)} reais')
print(f' faturamento da loja 1 na cadegoria de esporte_lazer:          {round(loja1_esporte_lazer,2)} reais')
print(f' faturamento da loja 1 na cadegoria de livros:                 {round(loja1_livros,2)} reais')

print('\n')

#media de faturamento por categoria
print('media de faturamento por categoria')
media_brinquedos=loja1_brinquedos/quantidade_brinquedos
media_eletronicos=loja1_eletronicos/quantidade_eletronicos
media_domesticas=loja1_utilidades_domesticas/quantidade_domesticas
media_musicais=loja1_instrumentos_musicais/quantidade_musicais
media_eletrodomesticos=loja1_eletrodomesticos/quantidade_eletrodomesticos
media_moveis=loja1_moveis/quantidade_moveis
media_esporte_lazer=loja1_esporte_lazer/quantidade_esporte_lazer
media_livros=loja1_livros/quantidade_livros
media_produtos= [media_domesticas,media_esporte_lazer,media_brinquedos,media_musicais,media_eletrodomesticos,media_eletronicos,media_moveis,media_livros]

i=0
while i<len(media_produtos):
  print(f'o preço medio de {categorias[i]}, e de  {round(media_produtos[i],2)} reais')
  i+=1


print('\n')


#graficos
faturamento_por_categoria=[loja1_utilidades_domesticas,loja1_esporte_lazer,loja1_brinquedos,loja1_instrumentos_musicais,loja1_eletrodomesticos,loja1_eletronicos,loja1_moveis,loja1_livros]
plt.title('faturamento por categoria da loja 1')
plt.pie(faturamento_por_categoria, labels=categorias,autopct='%1.1f%%')
plt.show()
print('')

plt.title('quantidade de produtos por categoria da loja 1')
plt.pie(quantidades, labels=categorias,autopct='%1.1f%%')
plt.show()








````
